```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
X,_ = make_blobs(n_samples=300,centers=4,cluster_std=3,random_state=42)
```

```python
plt.figure()
plt.scatter(X[:,0],X[:,1])
plt.title("Örnek Veri")
```

```python
linkage_methods = ["ward","single", "average","complete"]
# ward : küme içi varyansı (intra-cluster variance) minimize edecek şekilde" kümeleri birleştirir
# single : iki küme arasındaki mesafe, birbirine en yakın iki nokta arasındaki mesafe
# average : iki kğme arasındaki mesafe, kümeler arası tüm nokta çiftlerinin ortalama mesafesidir
# complete : iki küme arasındaki mesafe, birbirine en uzak iki nokta arasındaki mesafedir
plt.figure(figsize = (16,8))
for i , linkage_methods in enumerate(linkage_methods,1):
    model = AgglomerativeClustering(n_clusters=3,linkage=linkage_methods)
    
    cluster_labels = model.fit_predict(X)
    
    plt.subplot(2,4,i)
    plt.title(f"{linkage_methods.capitalize()}\nLinkage \nDendogram")
    dendrogram(linkage(X,method=linkage_methods),no_labels=True)
    plt.xlabel("Veri noktaları")
    plt.ylabel("Uzaklık")

    plt.subplot(2,4,i+4)
    plt.scatter(X[:,0],X[:,1],c = cluster_labels, cmap="viridis")
    plt.title(f"{linkage_methods.capitalize()}\nLinkage \nClustering")
    plt.xlabel("X")
    plt.ylabel("Y")
   
   
 # satırlar arası boşluk
plt.subplots_adjust(hspace=1)
# sütunlar arası boşluk
plt.subplots_adjust(wspace=0.7)

```
