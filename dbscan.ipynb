```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
```

```python
X, _ = make_circles(n_samples=1000,factor=0.5,noise=0.08,random_state=45)
# factor : İç daire, dış dairenin yarı çapında olur
# noise : noktaları biraz rastgele dağıtarak gerçekçi hale getirir
```

```python
plt.figure(figsize = (6,6))
plt.scatter(X[:,0],X[:,1])
```

```python
dbscan = DBSCAN(eps = 0.15,min_samples=15)
# eps = yarıçap
# min samples = min point
```

```python
cluster_labels = dbscan.fit_predict(X)
```

```python
plt.figure(figsize = (6,6))
plt.scatter(X[:,0],X[:,1],c=cluster_labels,cmap="viridis")
plt.title("DBSCAN SONUÇLARI")
```

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) algoritması, yoğunluk bazlı çalıştığı için küme (cluster) sayısını otomatik belirler; kullanıcıdan küme sayısı girilmesi gerekmez.
