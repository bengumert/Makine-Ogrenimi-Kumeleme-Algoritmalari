```python
# veri seti oluşturacağız
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
X,_ = make_blobs(n_samples=300,centers=4,cluster_std=0.6,random_state=42)
# n_samples = 300 adet veri
# centers = 4 farklı küme
# cluster_std = her kümenin yayılımı(standart sapması)
# random_state  aynı rastgele veriyi her seferinde oluşturmak için kullanılır. (RNG sabitlemesi - deterministic output)
```

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1])
plt.title("Örnke Veri")
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)                         
```

```python
labels = kmeans.labels_
# her noktanın ait olduğu küme
centers = kmeans.cluster_centers_
# küme merkezlerini al
```

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1],c=labels,cmap="viridis")
plt.scatter(centers[:, 0], centers[:, 1],c="red",marker = "x")
plt.title("K-Means Kümeleme")
```
